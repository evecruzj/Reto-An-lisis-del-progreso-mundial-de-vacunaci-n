# a. Extraer la información del archivo

```python
import pandas as pd
import numpy as np

datos = pd.read_csv("country_vaccinations.csv")
datos.head()
```

# b. Mostrar la estructura y tipos de datos

```python
datos["date"] = pd.to_datetime(datos["date"])
datos.info()
```

# c. Cantidad de vacunas aplicadas de cada compañía

```python
vacunas_compania = datos.groupby("vaccines")["total_vaccinations"].sum()
vacunas_compania
```

# d. Cantidad de vacunas aplicadas en todo el mundo

```python
total_mundo = datos["total_vaccinations"].sum()
print ("La cantidad de vacunas aplicadas en todo el mundo es: ",total_mundo)
```

# e. Calcular el promedio de vacunas aplicadas por país

```python
promedio_pais = datos.groupby("country")["total_vaccinations"].mean()
promedio_pais
```

# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21

```python
vacunas_29 = datos[datos["date"]== "2021-01-29"]
total_29 = vacunas_29["daily_vaccinations"].sum()
print("La cantidad de vacunas aplicadas el 29/01/21 fue: ",total_29)
```

# g. Crear un dataframe llamado conDiferencias

```python
conDiferencias = datos.copy()
conDiferencias["diferencias"]=conDiferencias["daily_vaccinations"]- conDiferencias["daily_vaccinations_raw"]
conDiferencias.head()
```

# h. Obtener el periodo de tiempo entre la fecha más reciente y la más antigua

```python
fecha_mayor = datos["date"].max()

fecha_menor = datos["date"].min()

diferencia = fecha_mayor - fecha_menor

print("La diferencia de tiempo es:", diferencia)
```

# i. Crear un dataframe llamado conCantidad

```python
conCantidad = datos.copy()

conCantidad["canVac"] = conCantidad["vaccines"].str.split(",").str.len()

conCantidad.head()
```

# j. Crear un dataframe llamado antes20

```python
antes20 = datos[datos["date"] < "2020-12-20"]

antes20
```

# k. Crear un dataframe llamado pfizer

```python
pfizer = datos[datos["vaccines"].str.contains("Pfizer", na=False)]

pfizer
```

# l. Guardar los dataframes en un archivo de Excel

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="conCantidad", index=False)
    antes20.to_excel(writer, sheet_name="antes20", index=False)
    pfizer.to_excel(writer, sheet_name="pfizer", index=False)
```
